Detecting and handling outliers

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(ds["Sales"],vert=False)
plt.show()
```

```python
plt.boxplot(ds["Profit"],vert=False)
plt.show()
```

```python
plt.boxplot(ds["Discount"],vert=False)
plt.show()
```

```python
plt.boxplot(ds["Quantity"],vert=False)
plt.show()
```

so from the above boxplots it is clear that there are few outliers in above columns ,we have to handle it to draw right insights

```python
ds1=(ds[ds["Sales"]>1.1])
ds1=ds1[ds1["Sales"]<8000]
ds1
```

```python
plt.boxplot(ds1["Sales"],vert=False)
plt.show()
```

```python
print(ds1[ds1["Sales"]>5500])
```

```python
ds1=ds1[ds1["Sales"]<5500]
```

```python
plt.boxplot(ds1["Sales"],vert=False)
plt.show()
```

    Now the sales column is free from outliers and we assume that now it is normally distributed

    We will now do the same for profit also

```python
ds[ds["Profit"]>2000]
```

```python
ds1=(ds[ds["Profit"]>-2000])
ds1=ds1[ds1["Profit"]<2500]
ds1.info()
```

```python
plt.boxplot(ds1["Profit"],vert=False)
plt.show()
```

    After removing outliers we can observe that the density of the products being sold more without profit and loss of amount of loss between 0 to 1300 approximately.
    We even observe that there are also more products sold at good amount of profit ranging from 0 to 1500.
    now its all about knowing the factors affecting the profit and the KPI of profit to maximize the sales with objective of increasing the profit and decresing the loss .

```python
plt.boxplot(ds1["Discount"],vert=False)
plt.show()
```

    Here we need to know about what count products we have on discount above 0.5

```python
ds1[ds1["Discount"]>0.2].describe()
```

```python
plt.scatter(ds1["Discount"],ds1["Profit"])
plt.show()
```

```python
plt.scatter(ds1["Sales"],ds1["Profit"])
plt.show()
```

    From the above plot we can observe that with discount more than 0.2 per product the supermarket is facing high loss rather than profit.
    The purpose or the strategy for applying discount might be for selling them asap before thier expiry or to sell them knowing the coming new market and new product models.
    So we should now try to know the reason of having products unsold for many days or the ones leading to application of discounts which is in turn reducing the profit and brings loss.
